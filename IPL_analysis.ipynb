# Outlier Detection On an Inning Score

here is the outlier got through analyzing the first inning score of the match btw RCB and Pune.

```python
match_df[(match_df['Team1']=='Royal Challengers Bengaluru') & (match_df['Team2']=='Rising Pune Supergiant') & (match_df['Season']=='2016')]['ID']
```

```python
outlier_df = delivery_df[
    (delivery_df['ID'] ==980969) & (delivery_df['innings'] == 1)
].copy()

outlier_df['ball'] = (outlier_df['overs'] - 1) * 6 + outlier_df['ballnumber']

outlier_df = outlier_df.sort_values('ball')

Q1 =outlier_df['total_run'].quantile(0.25)
Q3 = outlier_df['total_run'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_iqr =outlier_df[(outlier_df['total_run'] < lower) | (outlier_df['total_run'] > upper)]
outliers_iqr
```

The Z-score method did not identify any outliers using the threshold |z| > 3, indicating that no innings scores lie more than three standard deviations away from the mean. This suggests that the distribution does not contain extreme values.
However, the IQR method identified several outliers. This difference arises because the IQR method is more robust to skewed data and can detect moderate deviations from the median, whereas the Z-score method is sensitive to the mean and standard deviation.

```python
from scipy.stats import zscore
outlier_df['z_score'] = zscore(outlier_df['total_run'])
outlier=outlier_df[np.abs(outlier_df['z_score'])>3]
outlier
```

# Chi Square Test

Null Hypothesis: The match outcome is independent of the Toss Decision
Alternate Hypothesis: The match outcome has significally impact on the The Toss Decision

```python
match_df['toss_win_match_win']=(match_df['TossWinner']==match_df['WinningTeam'])
table=pd.crosstab(match_df['TossDecision'],match_df['toss_win_match_win'])
table
```

```python
from scipy.stats import chi2_contingency
chi2,p,dof,expected=chi2_contingency(table)
print(p)
print(chi2)
print(dof)
```

We reject the Null Hypothesis,since the p is less than alpha(the significant value) we can say that toss decision does have a statistically significant impact on the match outcome.

# Prediction of Team Batting First Win or Not(Using Match Level Features at the time of Toss)

Feature Engineering and Feature Selection

Here i have used some engineered historical features to imporve the accuracy of my model because it was not performing well using only raw data
some of the historical engineered feature i included are like Team's recent wins and overall both team rate difference Team who is batting first and difference of how many times both team won by batting first

```python
match_df.info()
```

here i tried to get features like the team recent form and the overall team rate of the batting first team and batting second team. For that first i made a separate dataframes for Team 1 and Team 2 with the Winning Team and then giving them same feature as Team i concatenated them so that i can get their proper win rate ensuring that the same team can occur as team1 as well as team2

```python
match_df['Date']=pd.to_datetime(match_df['Date'])
```

```python
team1_df = match_df[['Date','Team1','WinningTeam']].copy()
team1_df.columns = ['Date','Team','WinningTeam']

team2_df = match_df[['Date','Team2','WinningTeam']].copy()
team2_df.columns = ['Date','Team','WinningTeam']

long_df = pd.concat([team1_df, team2_df])
```

```python
long_df['win'] = (long_df['Team'] == long_df['WinningTeam']).astype(int)
long_df = long_df.sort_values('Date')
```

```python
long_df['matches_played'] = long_df.groupby('Team').cumcount()
long_df['cum_wins'] = (
    long_df.groupby('Team')['win']
    .transform(lambda x: x.cumsum().shift(1))
    .fillna(0)
)

long_df['win_rate'] = long_df['cum_wins'] / (long_df['matches_played'] + 1)
```

With the Win rate and the Win feature we can extract features like recent win and the win rate difference. And point to note that the Win rate is calculated only using the past matches data to avoid data leakage.

```python
long_df['recent_wins'] = (
    long_df.groupby('Team')['win']
    .transform(lambda x: x.shift(1).rolling(5).sum())
)
long_df['recent_matches'] = (
    long_df.groupby('Team').cumcount().clip(upper=5)
)
long_df['recent_wins'] = long_df['recent_wins'].fillna(0)
long_df['recent_win_rate'] = (
    long_df['recent_wins'] / (long_df['recent_matches'] + 1)
)
```

```python
team1_stats_long_df = long_df.iloc[:len(match_df)]
team2_stats_long_df = long_df.iloc[len(match_df):]

match_df['team1_rate'] = team1_stats_long_df['win_rate'].values
match_df['team2_rate'] = team2_stats_long_df['win_rate'].values
match_df['team1_rate_recent'] = team1_stats_long_df['recent_win_rate'].values
match_df['team2_rate_recent'] = team2_stats_long_df['recent_win_rate'].values
```

```python
match_df['win_rate_diff'] = match_df['team1_rate'] - match_df['team2_rate']
```

```python
match_df['batting_first_team'] = match_df.apply(
    lambda x: x['Team1'] if x['TossDecision']=='bat' and x['TossWinner']==x['Team1']
    else x['Team2'] if x['TossDecision']=='bat'
    else x['Team2'] if x['TossWinner']==x['Team1']
    else x['Team1'],
    axis=1
)
```

```python
match_df['bat_first_rate'] = np.where(
    match_df['batting_first_team'] == match_df['Team1'],
    match_df['team1_rate'],
    match_df['team2_rate']
)

match_df['bat_second_rate'] = np.where(
    match_df['batting_first_team'] == match_df['Team1'],
    match_df['team2_rate'],
    match_df['team1_rate']
)
```

```python
match_df['strength_diff'] = match_df['bat_first_rate'] - match_df['bat_second_rate']
```

```python
match_df=match_df.drop(['team1_rate','team2_rate','bat_first_rate','bat_second_rate'],axis=1)
```

```python
match_df['toss_bat'] = (match_df['TossDecision'] == 'bat').astype(int)
```

```python
match_df['toss_bat']
```

 we have to encode important categorical columns and columns which are not important can be dropped and also there are some columns which is decided after the match so we can also drop it.Matchnumber column is a type of mixed column so instead of relying on the number we can just form the separate column which has 3 or 4 categories such as playoff semifinal final and league matches

```python
#dropping features that have high cardinality and also some irrelevant features
match_df=match_df.drop(['Umpire1','Umpire2','Team2Players','Team1Players','ID','City','Season','Date','Player_of_Match','TossDecision'],axis=1)
#post match data and we have to decide at the time of the toss
match_df=match_df.drop(['WinningTeam','SuperOver','Margin'],axis=1)
def stage(x):
  if 'Final' in x:
    return 'Final'
  elif 'Qualifier 2' in x or 'Eliminator' in x or 'Qualifier 1' in x or 'Elimination Final' in x or 'Qualifier' in x or '3rd Place Play-Off' in x :
    return 'Playoff'
  elif 'Semi Final' in x:
    return 'Semi Final'
  else:
    return 'League'

match_df['MatchStages'] = match_df['MatchNumber'].astype(str).apply(stage)
match_df=match_df.drop('MatchNumber',axis=1)
match_df['MatchStages'].head(10)
```

 target column which is "did the team who won the toss and opt batting first won or not??"

```python
match_df['Target']=match_df['WonBy'].apply(lambda x : 1 if x=='Runs' else 0)
match_df=match_df.drop('WonBy',axis=1)
x=match_df.drop('Target',axis=1)
y=match_df['Target']
```

```python
match_df.info()
```

80/20 train-test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Using column transformer encoding categorical features and scaling numerical features whichever necessary

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
categorical_cols=['Team1','Team2','Venue','TossWinner','MatchStages','batting_first_team']
numerical_cols = ['strength_diff','team1_rate_recent', 'team2_rate_recent']
trans = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_cols),
        ('scaler', StandardScaler(), numerical_cols)
    ],
    force_int_remainder_cols=False
)
```

implementing logistic regression and training the model by passing through the pipeline

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe = Pipeline(steps = [
    ('step1', trans),
    ('step2', LogisticRegression())
])
```

Fitting and Evaluating the model

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay, precision_score, recall_score, f1_score , roc_curve,auc
pipe.fit(x_train,y_train)


```

```python
y_train_pred=pipe.predict(x_train)
train_accuracy=accuracy_score(y_train,y_train_pred)
```

```python
y_test_pred=pipe.predict(x_test)
test_accuracy=accuracy_score(y_test,y_test_pred)
```

```python
train_precision = precision_score(y_train, y_train_pred)
train_recall = recall_score(y_train, y_train_pred)
train_f1 = f1_score(y_train, y_train_pred)
```

```python
test_precision = precision_score(y_test, y_test_pred)
test_recall = recall_score(y_test, y_test_pred)
test_f1 = f1_score(y_test, y_test_pred)
```

```python
metrics_table = pd.DataFrame({
    'Dataset': ['Training', 'Test'],
    'Accuracy': [train_accuracy, test_accuracy],
    'Precision': [train_precision, test_precision],
    'Recall': [train_recall, test_recall],
    'F1 Score': [train_f1, test_f1]
})

print(metrics_table)
```

```python
metrics_table.to_csv("metrics_summary.csv", index=False)
```
